## Imports

```python
import pandas as pd
import csv
import numpy as np
import pickle
```

## Reading data

```python
dtype_list=pickle.load(open('type/train.pkl','rb'))
df=pd.read_csv('../../dataset/train_data_balance_todos.csv',dtype=dtype_list)
df_original=df.copy()
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
df
```

```python
def filtrar(df,col):
    df1=df[df.label==1]
    a=df1[col].value_counts().cumsum()/len(df1)
    lista=a[~(a>0.9)]
    return lista.index.values
```

## Missing Values 

```python
df.shape
```

```python
nullvalues=(df==-1).sum(axis=0).sort_values(ascending=False)
nullvalues.head(10)
```

```python
df.drop(['membership_life_duration','up_membership_grade','up_life_duration','his_on_shelf_time','his_app_size'],axis=1,inplace=True)
```

```python
df.drop(df[df.communication_avgonline_30d==-1].index,inplace=True)
df.drop(df[df.device_price==-1].index,inplace=True)
df.drop(df[df.list_time==-1].index,inplace=True)
df.drop(df[df.age==-1].index,inplace=True)
df.drop(df[df.city==-1].index,inplace=True)
```

```python
df.shape
```

```python
df.drop(['uid'],axis=1,inplace=True)
```

```python
df.drop(['task_id'],axis=1,inplace=True)
```

```python
df = df.drop('communication_onlinerate',axis=1)
```

## Dummies

```python
df = pd.get_dummies(df, columns=['gender'])
df = pd.get_dummies(df, columns=['city'])
df = pd.get_dummies(df, columns=['career'])
df = pd.get_dummies(df, columns=['residence'])
df = pd.get_dummies(df, columns=['device_name'])
df = pd.get_dummies(df, columns=['dev_id'])
df = pd.get_dummies(df, columns=['spread_app_id'])
df = pd.get_dummies(df, columns=['tags'])
df = pd.get_dummies(df, columns=['indu_name'])
df = pd.get_dummies(df, columns=['adv_prim_id'])
df = pd.get_dummies(df, columns=['inter_type_cd'])
df = pd.get_dummies(df, columns=['creat_type_cd'])

```

```python
adv_list=filtrar(df,'adv_id')
df.adv_id[~df.adv_id.isin(adv_list)] = -1
df = pd.get_dummies(df, columns=['adv_id'])
df.drop(['adv_id_-1'],axis=1,inplace=True)
```
