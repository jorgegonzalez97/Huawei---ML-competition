```python
import pandas as pd
import csv
import numpy as np
```

```python
path="dataset/train_data.csv"
df_sample=pd.read_csv(path, delimiter='|',nrows=10)
dict(zip(df_sample.columns,df_sample.dtypes))
```

```python
df_sample = df_sample.astype({'task_id': 'int16', 'adv_id':'int16', 'adv_prim_id':'int16', 'device_size':'int16', 'uid':'int32'})
```

```python
dtype_list = list()
for x in df_sample.dtypes.tolist():
    if x=='int16':
        dtype_list.append('int16')
    elif(x=='int32'):
        dtype_list.append('int32')
    elif(x=='int64'):
        dtype_list.append('int8')
    else:
        dtype_list.append('object')
        
dtype_list = dict(zip(df_sample.columns.tolist(),dtype_list))
dtype_list
```

```python
df = pd.read_csv(path, delimiter='|',dtype=dtype_list, nrows=30000000)
df
```

```python
df1=df[df.label==1]
df0=df[df.label==0]
print(len(df1))
print(len(df0))
```

```python
import random
random_vector = random.sample(range(len(df0)), len(df1))
random_vector.sort()
print(random_vector[0:10])
```

```python
df0=df0.iloc[random_vector]
df=df1.append(df0, ignore_index=True)
df=df.sample(frac=1)
df=df.reset_index()
df.drop(['index'],axis=1,inplace=True)
df
```

```python
df.to_csv('train_balanced.csv')
```
