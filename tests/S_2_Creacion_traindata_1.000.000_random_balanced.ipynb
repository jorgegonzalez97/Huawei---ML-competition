## Imports

```python
import pandas as pd
import csv
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Reading data

```python
#Path where the database is
path="../../dataset/train_data.csv"

#reading the database randomly
n = sum(1 for line in open(path)) - 1 #number of records in file (excludes header)
ROWS = 1000000 #desired sample size
skip = sorted(random.sample(range(1,n+1),n-ROWS)) #the 0-indexed header will not be included in the skip list
df=pd.read_csv(path, delimiter='|',nrows=ROWS,skiprows=skip)
df_original=df.copy()

pd.set_option('display.max_columns', None)
df
```

## Columns correction

```python
df['communication_onlinerate']=df['communication_onlinerate'].apply(lambda x:x.split('^'))
dummy=pd.get_dummies(df['communication_onlinerate'].apply(pd.Series).stack()).sum(level=0)
df = pd.concat([df, dummy], axis=1)
df = df.drop('communication_onlinerate',axis=1)
df.head()
```

```python
df['device_size_binned']=pd.cut(df['device_size'], 5,labels=False)
df['emui_dev_binned']=pd.cut(df['emui_dev'], 5,labels=False)
df['list_time_binned']=pd.cut(df['list_time'], 5,labels=False)
```
