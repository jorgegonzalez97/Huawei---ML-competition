```python
p=0.1

#Path where the database is
path="dataset/train_data.csv"

#reading the database
df = pd.read_csv(path,header=0, skiprows=lambda i: i>0 and random.random() > p,dtype=dtype_list,delimiter='|')
```

```python
df.label.value_counts()
```

```python
df0=df[df.label==0].sample(frac=1).iloc[0:len(df1)]
```

```python
del df 
```

```python
dffinal=df1.append(df0, ignore_index=True).sample(frac=1).reset_index(drop=True)
```
