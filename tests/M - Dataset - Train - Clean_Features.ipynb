## Imports

```python
import pandas as pd
import csv
import numpy as np
import pickle
```

## Reading data

```python
dtype_list=pickle.load(open('type/train.pkl','rb'))
df=pd.read_csv('../../dataset/train_data_balance.csv',dtype=dtype_list)
df_original=df.copy()
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
df
```

## Missing Values 

```python
df.shape
```

```python
nullvalues=(df==-1).sum(axis=0).sort_values(ascending=False)
nullvalues.head(10)
```

```python
df.drop(['membership_life_duration','up_membership_grade','up_life_duration','his_on_shelf_time','his_app_size'],axis=1,inplace=True)
```

```python
df.drop(df[df.communication_avgonline_30d==-1].index,inplace=True)
df.drop(df[df.device_price==-1].index,inplace=True)
df.drop(df[df.list_time==-1].index,inplace=True)
df.drop(df[df.age==-1].index,inplace=True)
```

```python
df.shape
```

```python
df.drop(['uid','adv_id','task_id'],axis=1,inplace=True)
```

```python
df = df.drop('communication_onlinerate',axis=1)
```

## Dummies

```python
df = pd.get_dummies(df, columns=['gender'])
df = pd.get_dummies(df, columns=['city'])
df = pd.get_dummies(df, columns=['career'])
df = pd.get_dummies(df, columns=['residence'])
df = pd.get_dummies(df, columns=['device_name'])
df = pd.get_dummies(df, columns=['dev_id'])
df = pd.get_dummies(df, columns=['spread_app_id'])
df = pd.get_dummies(df, columns=['tags'])
df = pd.get_dummies(df, columns=['indu_name'])
df = pd.get_dummies(df, columns=['adv_prim_id'])
df = pd.get_dummies(df, columns=['inter_type_cd'])
df = pd.get_dummies(df, columns=['creat_type_cd'])
df
```

## Bins 

```python
df['device_size_binned'],bins_device_size=pd.cut(df['device_size'], 5,labels=False,retbins=True)
df['device_size_binned'].value_counts()
```

```python
df['emui_dev_binned'],bins_emui_dev=pd.cut(df['emui_dev'], 5,labels=False,retbins=True)
df['emui_dev_binned'].value_counts()
```

```python
df['list_time_binned'],bins_list_time=pd.cut(df['list_time'], 5,labels=False,retbins=True)
df['list_time_binned'].value_counts()
```

```python
df['app_second_class_binned'],bins_app_second_class=pd.cut(df['app_second_class'], 5,labels=False,retbins=True)
df['app_second_class_binned'].value_counts()
```

```python
df.to_csv('../../dataset/train_data_clean_features.csv',index=False)
```

```python
import pickle

dtype_list = dict(zip(df.columns.tolist(),df.dtypes.tolist()))
pickle.dump(dtype_list,open('type/train_clean_features.pkl','wb'))

pickle.dump(bins_device_size,open('bins/device_size.pkl','wb'))
pickle.dump(bins_emui_dev,open('bins/emui_dev.pkl','wb'))
pickle.dump(bins_list_time,open('bins/list_time.pkl','wb'))
pickle.dump(bins_app_second_class,open('bins/app_second_class.pkl','wb'))
```
